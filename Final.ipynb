```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from IPython import display
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r'salaries_cyber.csv')
```

```python
#level of Exprince
plt.figure(figsize=(7,3))
plt.subplot(1,2,1)
df['experience_level'].value_counts().plot(kind='bar',color=sns.color_palette("tab10"))
plt.subplot(1,2,2)
df['experience_level'].value_counts().plot(kind='pie',autopct="%.1f")
plt.show()
```

```python
#salary depend on exprience
plt.figure(figsize=(10,4))
sns.displot(data = df['experience_level'], x = df["salary_in_usd"], hue = df["experience_level"], kind = "kde")
```

```python
#salary in every year by Exprience_level using bar plot and box plot
# in box plot easyly detect the outlier
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.barplot(x = df['work_year'], y = df['salary_in_usd'],hue = df['experience_level'])
plt.subplot(1,2,2)
sns.boxplot(x = df['work_year'], y = df['salary_in_usd'],hue = df['experience_level'])
plt.show()
plt.show()
```
